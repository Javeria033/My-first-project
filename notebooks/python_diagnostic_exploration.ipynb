## Task 1: Student Performance Analyzer

**Test Input:** `[78, 45, 89, 56, 32, 67, 91, 49]`

## Task 2: Grade Classification

**Grading Criteria:**
- A: 85-100
- B: 70-84
- C: 60-69
- D: 50-59
- F: Below 50

```python
test_scores = [92, 76, 65, 55, 42, -5, 105]
print("Testing various scores:")
for score in test_scores:
    try:
        grade = assign_grade(score)
        print(f"  Score {score} -> Grade: {grade}")
    except (ValueError, TypeError) as e:
        print(f"  Score {score} -> Error: {e}")
```

**Observation:** Invalid scores (-5, 105) are properly rejected with clear error messages.

## Task 3: Word Frequency Counter

```python
text = "Artificial intelligence is changing the world. Intelligence systems are becoming more powerful."
print(f"Input text: {text}")

freq, top3 = count_words(text)
print("Word Frequencies:")
for word, count in freq.items():
    print(f"  {word}: {count}")

print("\nTop 3 Most Frequent Words:")
for i, (word, count) in enumerate(top3, 1):
    print(f"  {i}. {word} ({count} times)")
```

**Observation:** "intelligence" appears 2 times (case-insensitive), and punctuation is correctly removed.

## Task 4: Data Cleaning Exercise

```python
student_records = [
    {"name": "huda", "score": 78},
    {"name": "haram", "score": None},
    {"name": "jhanzaib", "score": 92},
    {"name": "", "score": 65},
    {"name": "izhar", "score": 110},
    {"name": "ibtisam", "score": 48}
]

print(f"Original records ({len(student_records)} total):")
for record in student_records:
    print(f"  {record}")

cleaned, avg = clean_student_records(student_records)

print(f"\nCleaned records ({len(cleaned)} valid):")
for record in cleaned:
    print(f"  {record}")

print(f"\nAverage valid score: {avg}")
```

**Observation:** Only 3 out of 6 records are valid. Haram (None score), empty name, and Izhar (110 > 100) were correctly removed.

## Notebook vs Module Comparison

### What is a notebook useful for?
- **Exploration:** Testing ideas and experimenting with code interactively
- **Visualization:** Displaying charts, tables, and outputs inline
- **Documentation:** Adding markdown explanations alongside code
- **Prototyping:** Quickly building and testing functions before finalizing

### What is a Python module useful for?
- **Reusability:** Functions can be imported and reused across projects
- **Production:** Clean, organized code suitable for deployment
- **Testing:** Easier to write automated tests for modules
- **Version Control:** Better diff support and cleaner Git history

### Why should production logic not remain only inside a notebook?
1. **Reproducibility Issues:** Notebooks can have hidden state and execution order problems
2. **Version Control:** Notebooks are hard to diff and review in Git
3. **Testing:** Automated testing is difficult with notebook cells
4. **Deployment:** Notebooks cannot be easily deployed as APIs or services

### When would an AI professional use both?
- **Development Phase:** Use notebooks for EDA, prototyping, and visualization
- **Production Phase:** Move finalized code to modules for deployment
- **Reporting:** Use notebooks to create reports and share findings with stakeholders
